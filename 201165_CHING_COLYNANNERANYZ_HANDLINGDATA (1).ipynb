```python
#CODE CELL 
#PROBLEM 2
def tic_tac_toe (board): 
    
    try:     
        #vertically 
        for vertical in [vertical for vertical in zip (*board)]: 
             while len(set(vertical))==1: 
                return vertical[0]

        #horizontally 
        for horizontal in board: 
            while len(set(horizontal))==1: 
                return horizontal[0]

        #updown diagonals 
        if all([updown == "X" for updown in [board [i][i] for i,v in enumerate (board)]]): 
            return "X"

        elif all ([updown == "O" for updown in [board [i][i] for i,v in enumerate (board)]]):
            return "O" 

        #downup diagonals 
        elif all ([downup == "X" for downup in [board [2-i] [i] for i,v in enumerate (board)]]): 
            return "X"

        elif all ([downup == "O" for downup in [board [2-1][i] for i,v in enumerate (board)]]): 
            return "O"

        #none 
        else: 
            return ("None")

    except: 
        return ("None")

tic_tac_toe(board4)
    
```
